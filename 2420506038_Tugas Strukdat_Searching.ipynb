```python
import random
import math

def generate_sorted_list():
    sorted_list = sorted(random.sample(range(1, 1001), 100))
    print("Generated Sorted Array:", sorted_list)
    return sorted_list

def linear_search(arr, target):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            return i, comparisons
    return -1, comparisons

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    comparisons = 0

    while low <= high:
        mid = (low + high) // 2
        comparisons += 1
        if arr[mid] == target:
            return mid, comparisons
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, comparisons

def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0
    comparisons = 0

    while prev < n and arr[min(step, n) - 1] < target:
        comparisons += 1
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1, comparisons

    for i in range(prev, min(step, n)):
        comparisons += 1
        if arr[i] == target:
            return i, comparisons

    return -1, comparisons

data = generate_sorted_list()

target = random.choice(data)
print(f"Target: {target}")

# Menghitung perbandingan untuk tiap algoritma
linear_comparisons = linear_search(data, target)
binary_comparisons = binary_search(data, target)
jump_comparisons = jump_search(data, target)

# Menampilkan hasil
print("\nHasil Perbandingan:")
print("Algoritma Pencarian : Jumlah Perbandingan")
print(f"Linear Search : {linear_comparisons}")
print(f"Binary Search : {binary_comparisons}")
print(f"Jump Search : {jump_comparisons}")

# Menentukan algoritma yang paling efisien
results = {
    "Linear Search": linear_comparisons,
    "Binary Search": binary_comparisons,
    "Jump Search": jump_comparisons
}
paling_efisien = min(results, key=results.get)
print(f"\nAlgoritma paling efisien: {paling_efisien} ({results[paling_efisien]} perbandingan)")
```
