```python
class Satpam:
    def __init__(self, nama):
        self.nama = nama
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def tambah_satpam(self, nama):
        satpam_baru = Satpam(nama)
        if self.head is None:
            self.head = satpam_baru
            satpam_baru.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = satpam_baru
            satpam_baru.next = self.head

    def tampilkan_jadwal(self, jumlah_hari):
        if self.head is None:
            print("Tidak ada satpam yang terdaftar.")
            return

        print("\n\U0001F4A1 Jadwal Piket Satpam Selama 2 Minggu:")
        temp = self.head
        for hari in range(1, jumlah_hari + 1):
            print(f"Hari {hari}: {temp.nama}")
            temp = temp.next

def main():
    jadwal = CircularLinkedList()

    for i in range(1, 5):
        nama = input(f"Masukkan nama satpam ke-{i}: ")
        jadwal.tambah_satpam(nama)

    jadwal.tampilkan_jadwal(14)

if __name__ == "__main__":
    main()
```
