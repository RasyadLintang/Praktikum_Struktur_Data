# **Latihan/Tugas 1**

```python
class SistemAntrianBioskop:
    def __init__(self, kapasitas=10):
        self.antrian = []
        self.kapasitas_antrian = kapasitas

    def enqueue(self, ticket_id):
        if '7' in ticket_id:
            print(f"Pelanggan VIP {ticket_id} langsung mendapatkan kursi tanpa mengantri!")
            return True

        if self.size() >= self.kapasitas_antrian:
            print(f"Maaf, antrian sudah penuh. Pelanggan {ticket_id} harus menunggu hingga ada slot kosong.")
            return False

        self.antrian.append(ticket_id)
        print(f"Pelanggan {ticket_id} berhasil ditambahkan ke antrian.")
        return True

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong. Tidak ada pelanggan yang dapat dilayani.")
            return None

        pelanggan = self.antrian.pop(0)
        print(f"Pelanggan {pelanggan} sedang dilayani.")
        return pelanggan

    def front(self):
        if self.is_empty():
            print("Antrian kosong. Tidak ada pelanggan pertama.")
            return None

        return self.antrian[0]

    def rear(self):
        if self.is_empty():
            print("Antrian kosong. Tidak ada pelanggan terakhir.")
            return None

        return self.antrian[-1]

    def is_empty(self):
        return len(self.antrian) == 0

    def size(self):
        return len(self.antrian)

    def display_queue(self):
        if self.is_empty():
            print("Antrian kosong.")
            return

        print("Daftar Pelanggan dalam Antrian:")
        for i, pelanggan in enumerate(self.antrian, 1):
            print(f"{i}. {pelanggan}")

def main():
    sistem = SistemAntrianBioskop(10)

    while True:
        print("\nSISTEM ANTRIAN PEMESANAN TIKET BIOSKOP")
        print("1. Tambah Pelanggan")
        print("2. Layani Pelanggan")
        print("3. Lihat Pelanggan Pertama")
        print("4. Lihat Pelanggan Terakhir")
        print("5. Cek Status Antrian")
        print("6. Tampilkan Semua Antrian")
        print("0. Keluar")

        try:
            pilihan = int(input("Pilih menu: "))
        except ValueError:
            print("Masukkan angka yang valid!")
            continue

        if pilihan == 1:
            nama = input("Masukkan nama mahasiswa: ")
            npm = input("Masukkan NPM: ")

            kode_pelanggan = ""
            if len(nama) >= 2:
                kode_pelanggan = nama[:2].upper()
            else:
                kode_pelanggan = nama.upper()

            if len(npm) >= 2:
                kode_pelanggan += npm[-2:]

            sistem.enqueue(kode_pelanggan)

        elif pilihan == 2:
            sistem.dequeue()

        elif pilihan == 3:
            first = sistem.front()
            if first:
                print(f"Pelanggan pertama dalam antrian: {first}")

        elif pilihan == 4:
            last = sistem.rear()
            if last:
                print(f"Pelanggan terakhir dalam antrian: {last}")

        elif pilihan == 5:
            print("Status antrian:")
            print(f"- Jumlah pelanggan: {sistem.size()}")
            print(f"- Kapasitas maksimal: {sistem.kapasitas_antrian}")
            print(f"- Antrian kosong: {'Ya' if sistem.is_empty() else 'Tidak'}")

        elif pilihan == 6:
            sistem.display_queue()

        elif pilihan == 0:
            print("Terima kasih telah menggunakan sistem antrian!")
            break

        else:
            print("Pilihan tidak valid. Silakan coba lagi.")


if __name__ == "__main__":
    main()
```

# **Latihan/Tugas 2**

```python
class QueueNode:
    def __init__(self, order_id, num_pages):
        self.order_id = order_id
        self.num_pages = num_pages
        self.next = None

class FotokopiQueue:
    def __init__(self):
        self.front = None
        self.rear = None
        self.count = 0

    def is_empty(self):
        return self.front is None

    def size(self):
        return self.count

    def enqueue(self, order_id, num_pages):
        new_node = QueueNode(order_id, num_pages)

        if num_pages > 100:
            print(f"Pesanan {order_id} dengan {num_pages} halaman akan diproses dalam dua tahap")

        if num_pages < 10:
            print(f"Pesanan {order_id} masuk Fast Track! Diproses langsung tanpa antri")
            return

        if self.is_empty():
            self.front = new_node
            self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

        self.count += 1
        print(f"Pesanan {order_id} dengan {num_pages} halaman ditambahkan ke antrian")

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong, tidak ada pesanan yang dapat dilayani")
            return None

        temp = self.front
        order_id = temp.order_id
        num_pages = temp.num_pages

        self.front = self.front.next
        if self.front is None:
            self.rear = None

        self.count -= 1
        print(f"Melayani pesanan {order_id} dengan {num_pages} halaman")
        return order_id, num_pages

    def front_order(self):
        if self.is_empty():
            print("Antrian kosong, tidak ada pesanan di depan")
            return None

        print(f"Pesanan di depan: {self.front.order_id} dengan {self.front.num_pages} halaman")
        return self.front.order_id, self.front.num_pages

    def rear_order(self):
        if self.is_empty():
            print("Antrian kosong, tidak ada pesanan di belakang")
            return None

        print(f"Pesanan di belakang: {self.rear.order_id} dengan {self.rear.num_pages} halaman")
        return self.rear.order_id, self.rear.num_pages

    def display_all(self):
        if self.is_empty():
            print("Antrian kosong!")
            return

        print("\nDaftar Pesanan Fotokopi:")
        temp = self.front
        position = 1

        while temp:
            print(f"{position}. ID: {temp.order_id}, Jumlah halaman: {temp.num_pages}")
            temp = temp.next
            position += 1

def create_order_id(nama, npm):
    huruf_nama = nama[:3].upper()
    digit_npm = npm[-2:]
    return huruf_nama + digit_npm

def main():
    antrian_fotokopi = FotokopiQueue()

    print("SISTEM ANTRIAN LAYANAN FOTOKOPI KAMPUS : ")

    while True:
        print("\nMenu:")
        print("1. Tambah Pesanan")
        print("2. Layani Pesanan")
        print("3. Lihat Pesanan Pertama")
        print("4. Lihat Pesanan Terakhir")
        print("5. Cek Status Antrian")
        print("6. Tampilkan Semua Pesanan")
        print("0. Keluar")

        pilihan = input("\nPilih menu (0-6): ")

        if pilihan == "1":
            nama = input("Masukkan nama mahasiswa: ")
            npm = input("Masukkan NPM mahasiswa: ")
            jumlah_halaman = int(input("Masukkan jumlah halaman: "))

            order_id = create_order_id(nama, npm)
            antrian_fotokopi.enqueue(order_id, jumlah_halaman)

        elif pilihan == "2":
            antrian_fotokopi.dequeue()

        elif pilihan == "3":
            antrian_fotokopi.front_order()

        elif pilihan == "4":
            antrian_fotokopi.rear_order()

        elif pilihan == "5":
            if antrian_fotokopi.is_empty():
                print("Status: Antrian kosong")
            else:
                print(f"Status: Terdapat {antrian_fotokopi.size()} pesanan dalam antrian")

        elif pilihan == "6":
            antrian_fotokopi.display_all()

        elif pilihan == "0":
            print("Terima kasih telah menggunakan sistem antrian fotokopi!")
            break

        else:
            print("Pilihan tidak valid! Silakan coba lagi.")


if __name__ == "__main__":
    main()
```

# **Latihan/Tugas 3**

```python
class QueueNode:
    def __init__(self, vehicle_number, is_eco_friendly=False):
        self.vehicle_number = vehicle_number
        self.is_eco_friendly = is_eco_friendly
        self.next = None

class ParkingQueue:
    def __init__(self, capacity=15):
        self.front = None
        self.rear = None
        self.count = 0
        self.capacity = capacity

    def is_empty(self):
        return self.front is None

    def size(self):
        return self.count

    def is_full(self):
        return self.count >= self.capacity

    def enqueue(self, vehicle_number, is_eco_friendly=False):
        new_node = QueueNode(vehicle_number, is_eco_friendly)

        if self.is_full():
            print(f"Parkiran penuh! Kendaraan {vehicle_number} harus menunggu.")
            return False

        if is_eco_friendly:
            if self.is_empty():
                self.front = new_node
                self.rear = new_node
            else:
                if self.front.is_eco_friendly:
                    current = self.front
                    while current.next and current.next.is_eco_friendly:
                        current = current.next

                    new_node.next = current.next
                    current.next = new_node
                    if new_node.next is None:
                        self.rear = new_node
                else:
                    new_node.next = self.front
                    self.front = new_node

            print(f"Kendaraan {vehicle_number} masuk parkiran dengan prioritas (ramah lingkungan).")
        else:
            if self.is_empty():
                self.front = new_node
                self.rear = new_node
            else:
                self.rear.next = new_node
                self.rear = new_node

            print(f"Kendaraan {vehicle_number} masuk parkiran.")

        self.count += 1
        return True

    def dequeue(self):
        if self.is_empty():
            print("Parkiran kosong, tidak ada kendaraan yang dapat keluar")
            return None

        temp = self.front
        vehicle_number = temp.vehicle_number
        is_eco_friendly = temp.is_eco_friendly

        self.front = self.front.next
        if self.front is None:
            self.rear = None

        self.count -= 1
        if is_eco_friendly:
            print(f"Kendaraan {vehicle_number} (ramah lingkungan) keluar dari parkiran.")
        else:
            print(f"Kendaraan {vehicle_number} keluar dari parkiran.")

        return vehicle_number

    def front_vehicle(self):
        if self.is_empty():
            print("Parkiran kosong, tidak ada kendaraan di depan")
            return None

        if self.front.is_eco_friendly:
            print(f"Kendaraan di depan: {self.front.vehicle_number} (ramah lingkungan)")
        else:
            print(f"Kendaraan di depan: {self.front.vehicle_number}")

        return self.front.vehicle_number

    def rear_vehicle(self):
        if self.is_empty():
            print("Parkiran kosong, tidak ada kendaraan di belakang")
            return None

        if self.rear.is_eco_friendly:
            print(f"Kendaraan di belakang: {self.rear.vehicle_number} (ramah lingkungan)")
        else:
            print(f"Kendaraan di belakang: {self.rear.vehicle_number}")

        return self.rear.vehicle_number

    def display_all(self):
        if self.is_empty():
            print("Parkiran kosong!")
            return

        print("\nDaftar Kendaraan dalam Parkiran:")
        temp = self.front
        position = 1

        while temp:
            status = "(ramah lingkungan)" if temp.is_eco_friendly else ""
            print(f"{position}. Plat: {temp.vehicle_number} {status}")
            temp = temp.next
            position += 1
        print(f"Total: {self.count} kendaraan, Kapasitas: {self.capacity}")

def create_vehicle_number(kota_lahir, npm):
    huruf_kota = kota_lahir[:2].upper()
    digit_npm = npm[-2:]
    return huruf_kota + digit_npm

def main():
    parkiran_mall = ParkingQueue(capacity=15)

    print("SISTEM ANTRIAN LAYANAN PARKIR MALL")

    while True:
        print("\nMenu:")
        print("1. Tambah Kendaraan Masuk")
        print("2. Keluarkan Kendaraan")
        print("3. Lihat Kendaraan Pertama")
        print("4. Lihat Kendaraan Terakhir")
        print("5. Cek Status Parkiran")
        print("6. Tampilkan Semua Kendaraan")
        print("0. Keluar")

        pilihan = input("\nPilih menu (0-6): ")

        if pilihan == "1":
            kota_lahir = input("Masukkan kota lahir: ")
            npm = input("Masukkan NPM: ")
            ramah_lingkungan = input("Apakah kendaraan ramah lingkungan? (y/n): ").lower() == 'y'

            plat_nomor = create_vehicle_number(kota_lahir, npm)
            parkiran_mall.enqueue(plat_nomor, ramah_lingkungan)

        elif pilihan == "2":
            parkiran_mall.dequeue()

        elif pilihan == "3":
            parkiran_mall.front_vehicle()

        elif pilihan == "4":
            parkiran_mall.rear_vehicle()

        elif pilihan == "5":
            if parkiran_mall.is_empty():
                print("Status: Parkiran kosong")
            elif parkiran_mall.is_full():
                print(f"Status: Parkiran penuh ({parkiran_mall.size()} kendaraan)")
            else:
                print(f"Status: Tersedia {parkiran_mall.capacity - parkiran_mall.size()} slot dari {parkiran_mall.capacity} kapasitas")
                print(f"Jumlah kendaraan saat ini: {parkiran_mall.size()}")

        elif pilihan == "6":
            parkiran_mall.display_all()

        elif pilihan == "0":
            print("Terima kasih telah menggunakan sistem parkiran mall!")
            break

        else:
            print("Pilihan tidak valid! Silakan coba lagi.")


if __name__ == "__main__":
    main()
```

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class QueueLinkedList:
  def __init__(self):
    self.front = self.rear = None

  def enqueue(self, data):
    new_node = Node(data)
    if self.rear is None:
      self.front = self.rear = new_node
      return
    self.rear.next = new_node
    self.rear = new_node

  def dequeue(self):
    if self.front is None:
      return None
    dequeued_element = self.front.data
    self.front = self.front.next
    if self.front is None:
      self.rear = None
    return dequeued_element

  def is_empty(self):
    return self.front is None

  def front_element(self):
    return self.front.data if self.is_empty() else None

  def rear_element(self):
    return self.rear.data if self.is_empty() else None

# Contoh penggunaan
queue = QueueLinkedList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front_element())
print("Rear element:", queue.rear_element())
```

# **Latihan/Tugas 4**

```python
class AntrianSertifikat:
    def __init__(self):
        self.queue = []

    def enqueue(self, student_id):
        self.queue.append(student_id)

    def dequeue(self):
        if self.is_empty():
            return None
        return self.queue.pop(0)

    def front(self):
        if self.is_empty():
            return None
        return self.queue[0]

    def rear(self):
        if self.is_empty():
            return None
        return self.queue[-1]

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)


class SistemAntrianSertifikat:
    def __init__(self):
        self.antrian_reguler = AntrianSertifikat()
        self.antrian_berprestasi = AntrianSertifikat()
        self.mahasiswa_berprestasi = set()

    def enqueue(self, student_id):
        if student_id in self.mahasiswa_berprestasi:
            self.antrian_berprestasi.enqueue(student_id)
            print(f"Mahasiswa {student_id} (berprestasi) ditambahkan ke antrian khusus.")
        else:
            if self.antrian_reguler.size() >= 20:
                print("Antrian reguler melebihi 20 orang. Membagi menjadi dua jalur...")
                reguler_size = self.antrian_reguler.size()
                for _ in range(reguler_size // 2):
                    mahasiswa = self.antrian_reguler.dequeue()
                    self.antrian_berprestasi.enqueue(mahasiswa)
                print(f"Sekarang terdapat {self.antrian_reguler.size()} mahasiswa di jalur reguler dan {self.antrian_berprestasi.size()} mahasiswa di jalur berprestasi.")
            self.antrian_reguler.enqueue(student_id)
            print(f"Mahasiswa {student_id} ditambahkan ke antrian reguler.")

    def dequeue(self):
        if not self.antrian_berprestasi.is_empty():
            mahasiswa = self.antrian_berprestasi.dequeue()
            print(f"Melayani mahasiswa {mahasiswa} dari jalur berprestasi.")
            return mahasiswa
        elif not self.antrian_reguler.is_empty():
            mahasiswa = self.antrian_reguler.dequeue()
            print(f"Melayani mahasiswa {mahasiswa} dari jalur reguler.")
            return mahasiswa
        else:
            print("Antrian kosong. Tidak ada mahasiswa yang dapat dilayani.")
            return None

    def front(self):
        if not self.antrian_berprestasi.is_empty():
            return self.antrian_berprestasi.front()
        else:
            return self.antrian_reguler.front()

    def rear(self):
        if self.antrian_reguler.is_empty():
            return self.antrian_berprestasi.rear()
        else:
            return self.antrian_reguler.rear()

    def is_empty(self):
        return self.antrian_reguler.is_empty() and self.antrian_berprestasi.is_empty()

    def size(self):
        return self.antrian_reguler.size() + self.antrian_berprestasi.size()

    def tambah_mahasiswa_berprestasi(self, student_id):
        self.mahasiswa_berprestasi.add(student_id)
        print(f"Mahasiswa {student_id} ditambahkan ke daftar mahasiswa berprestasi.")

    def tampilkan_status_antrian(self):
        print("\nSTATUS ANTRIAN SERTIFIKAT")
        print(f"Jumlah total mahasiswa dalam antrian: {self.size()}")

        print("\nAntrian Jalur Berprestasi:")
        if self.antrian_berprestasi.is_empty():
            print("  [Kosong]")
        else:
            for i, student_id in enumerate(self.antrian_berprestasi.queue):
                print(f"  {i+1}. {student_id}")

        print("\nAntrian Jalur Reguler:")
        if self.antrian_reguler.is_empty():
            print("  [Kosong]")
        else:
            for i, student_id in enumerate(self.antrian_reguler.queue):
                print(f"  {i+1}. {student_id}")

def main():
    sistem = SistemAntrianSertifikat()

    print("SISTEM ANTRIAN PENGAMBILAN SERTIFIKAT")
    print("Format ID mahasiswa: [huruf pertama program studi] + [dua digit terakhir NPM]")
    print("Contoh: TI22 untuk mahasiswa Teknologi Informasi dengan NPM berakhiran 22")

    while True:
        print("\nMenu:")
        print("1. Tambah mahasiswa ke antrian")
        print("2. Layani mahasiswa")
        print("3. Lihat mahasiswa pertama")
        print("4. Lihat mahasiswa terakhir")
        print("5. Cek apakah antrian kosong")
        print("6. Hitung jumlah mahasiswa dalam antrian")
        print("7. Tambah mahasiswa berprestasi")
        print("8. Tampilkan status antrian")
        print("0. Keluar")

        pilihan = input("Pilih menu (0-8): ")

        if pilihan == "1":
            student_id = input("Masukkan ID mahasiswa: ")
            sistem.enqueue(student_id)

        elif pilihan == "2":
            sistem.dequeue()

        elif pilihan == "3":
            mahasiswa = sistem.front()
            if mahasiswa:
                print(f"Mahasiswa pertama dalam antrian: {mahasiswa}")
            else:
                print("Antrian kosong.")

        elif pilihan == "4":
            mahasiswa = sistem.rear()
            if mahasiswa:
                print(f"Mahasiswa terakhir dalam antrian: {mahasiswa}")
            else:
                print("Antrian kosong.")

        elif pilihan == "5":
            if sistem.is_empty():
                print("Antrian kosong.")
            else:
                print("Antrian tidak kosong.")

        elif pilihan == "6":
            print(f"Jumlah mahasiswa dalam antrian: {sistem.size()}")

        elif pilihan == "7":
            student_id = input("Masukkan ID mahasiswa berprestasi: ")
            sistem.tambah_mahasiswa_berprestasi(student_id)

        elif pilihan == "8":
            sistem.tampilkan_status_antrian()

        elif pilihan == "0":
            print("Terima kasih telah menggunakan sistem antrian.")
            break

        else:
            print("Pilihan tidak valid. Silakan coba lagi.")


if __name__ == "__main__":
    main()
```
