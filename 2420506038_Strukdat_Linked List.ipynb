```python
# Definisi Node dalam Linked List
class TaskNode:
  def __init__ (self, task_name):
    self.task_name = task_name # Menyimpan nama tugas
    self.next = None # Pointer ke tugas berikutnya
# Definisi Linked List untuk Daftar Tugas
class TaskList:
  def __init__ (self):
    self.head = None # Awalnya tidak ada tugas (daftar kosong)
# Menambahkan tugas ke Linked List
  def add_task(self, task_name):
    new_task = TaskNode (task_name)
    if not self.head:
      self.head =new_task # Jika list kosong, tugas pertama menjadi head

    else:
      current = self.head
      while current.next: #Mencari node terakhir dalam list
        current = current.next
      current.next = new_task # Menambahkan tugas di akhir list
# Menampilkan daftar tugas
  def show_tasks(self):
    if not self.head:
      print("\n Tidak ada mata kuliah hari Jumat!")
      return

    print("\n Daftar Mata Kuliah Hari Jumat:")
    current = self.head
    count = 1
    while current:
      print(f"{count}. {current.task_name}")
      current = current.next
      count += 1
#Program Utama (Main)
# Membuat objek daftar tugas
task_list = TaskList()
#
# Meminta pengguna memasukkan jumlah tugas
num_tasks = int (input ("Masukkan jumlah mata kuliah hari jumat: ") )
for i in range(1, num_tasks + 1):
  task_name = input (f"Masukkan mata kuliah ke-{i}: ")
  task_list.add_task(task_name)
task_list.show_tasks()
```

```python
class PageNode:
  def __init__(self, url):
    self.url = url
    self.prev = None
    self.next = None

class BrowserHistory:
  def __init__(self):
    self.current = None

  def visit_page(self, url):
      new_page = PageNode(url)
      if not self.current:
        self.current = new_page
      else:
        new_page.prev = self.current
        self.current.next = new_page
        self.current = new_page

  def show_history(self):
      if not self.current:
        print("\n Riwayat kosong.")
        return

      print("\nRiwayat Peramban Web:")
      temp = self.current
      while temp:
        temp = temp.prev

      count = 1
      while temp:
        print(f"{count}. {temp.url}")
        temp = temp.next
        count += 1

  def go_back(self):
      if self.current and self.current.prev:
        self.current = self.current.prev
        print(f"\n Anda kembali ke: {self.current.url}")
      else:
        print("\n Tidak bisa kemnbali, ini adalah halaman pertama!.")

  def go_forward(self):
      if self.current and self.current.next:
        self.current = self.current.next
        print(f"\n Anda maju ke: {self.current.url}")
      else:
        print("\n Tidak bisa maju, ini adalah halaman terakhir!.")

history = BrowserHistory()

num_pages = int(input("Masukkan jumlah halaman yang ingin Anda kunjungi: "))

for i in range(1, num_pages + 1):
    url = input(f"Masukkan URL halaman ke-{i}: ")
    history.visit_page(url)

history.show_history()

while True:
  command = input("\nMasukkan perintah (back/forward/exit): ").lower()

  if command == "back":
    history.go_back()
  elif command == "forward":
    history.go_forward()
  elif command == "exit":
    print("Keluar dari riwayat peramban.")
    break
  else:
    print("Perintah tidak valid")
```

```python
# Definisi Node dalam Circular Linked List
class SongNode:
    def __init__(self, title):
        self.title = title  # Menyimpan judul lagu
        self.next = None  # Pointer ke lagu berikutnya

# Definisi Circular Linked List untuk Playlist Lagu
class Playlist:
    def __init__(self):
        self.current = None  # Lagu yang sedang diputar

    # Menambahkan lagu ke dalam playlist
    def add_song(self, title):
        new_song = SongNode(title)
        if not self.current:
            self.current = new_song
            self.current.next = self.current  # Menghubungkan ke dirinya sendiri (Circular)
        else:
            temp = self.current
            while temp.next != self.current:  # Mencari lagu terakhir
                temp = temp.next
            temp.next = new_song  # Menambahkan lagu di akhir
            new_song.next = self.current  # Menghubungkan kembali ke awal

    # Menampilkan daftar lagu dalam playlist
    def show_playlist(self):
        if not self.current:
            print("\nPlaylist kosong!")
            return

        print("\nPlaylist Lagu Anda:")
        temp = self.current
        count = 1
        while True:
            print(f"{count}. {temp.title}")
            temp = temp.next
            count += 1
            if temp == self.current:  # Berhenti jika sudah kembali ke awal
                break

    # Memutar lagu berikutnya dalam playlist
    def next_song(self):
        if self.current:
            self.current = self.current.next
            print(f"\nMemutar: {self.current.title}")

# ====== Program Utama (Main) ====== #

# Membuat objek playlist
playlist = Playlist()

# Meminta pengguna memasukkan jumlah lagu
num_songs = int(input("\nMasukkan jumlah lagu dalam playlist: "))

# Loop untuk memasukkan lagu ke dalam Circular Linked List
for i in range(1, num_songs + 1):
    title = input(f"Masukkan judul lagu ke-{i}: ")
    playlist.add_song(title)

# Menampilkan daftar lagu
playlist.show_playlist()

# Menu navigasi pengguna
while True:
    command = input("\nTekan 'n' untuk lagu berikutnya, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'n':
        playlist.next_song()
    elif command == 'x':
        print("\nKeluar dari pemutar musik.")
        break
    else:
        print("\nPerintah tidak dikenal.")
```

```python
# Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operasi
class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, data, position):
        if position < 0:
            print("Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print("Posisi melebihi panjang daftar!")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    # Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next

    # Menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    # Menghapus elemen di posisi tertentu
    def delete_at_position(self, position):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.head = self.head.next
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp.next:
                print("Posisi tidak ditemukan!")
                return
            temp = temp.next
        temp.next = temp.next.next

    # Mencari elemen dalam Linked List
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f"Elemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print(f"Elemen {key} tidak ditemukan dalam daftar.")

    # Membalik (reverse) Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display(self):
        temp = self.head
        if not temp:
            print("Daftar kosong!")
            return
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")


# ===== Program Utama (Main) ===== #

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("Masukkan jumlah elemen awal elemen dalam daftar: "))
for i in range(num_elements):
    data = int(input(f"Masukkan data ke-{i+1}: "))
    ll.add_at_end(data)

# Menu Operasi Linked List
while True:
    print("\nPilih Operasi:")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi tertentu")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balikkan data")
    print("9. Tampilkan daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data yang ingin ditambahkan: "))
        ll.add_at_position(data, position)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print("Daftar berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\nKeluar dari program.")
        break
    else:
        print("\nPilihan tidak valid, silakan coba lagi!")
```

```python
class Node:
    def __init__(self, name):
        self.name = name
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, name):
        new_node = Node(name)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def get_schedule(self, days):
        schedule = []
        temp = self.head
        for _ in range(days):
            schedule.append(temp.name)
            temp = temp.next
        return schedule

# Input nama satpam
satpam_list = []
for i in range(4):
    name = input(f"Masukkan nama satpam ke-{i+1}: ")
    satpam_list.append(name)

# Buat circular linked list
cll = CircularLinkedList()
for name in satpam_list:
    cll.append(name)

# Cetak jadwal selama 14 hari
print("\n\U0001F4A1 Jadwal Piket Satpam Selama 2 Minggu:")
for day, name in enumerate(cll.get_schedule(14), start=1):
    print(f"Hari {day}: {name}")
```
